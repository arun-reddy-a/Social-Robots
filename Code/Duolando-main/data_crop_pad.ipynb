```python
import numpy as np
import pickle
import os
from tqdm import tqdm
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
```

```python
folder_path = './data/motion/pos3d/train'
```

```python
train_file_names = (sorted([f for f in os.listdir(folder_path) if f.endswith('.npy')]))
```

```python
frame_count = 0
```

```python
for i in tqdm(range(len(train_file_names)//2)):
    follower_file = np.load(os.path.join(folder_path, train_file_names[i*2]))
    leader_file = np.load(os.path.join(folder_path, train_file_names[i*2+1]))
    frame_count += len(follower_file)
```

```python
train_result = []
```

```python
for i in tqdm(range(len(train_file_names)//2)):
    follower_file = np.load(os.path.join(folder_path, train_file_names[i*2]))
    leader_file = np.load(os.path.join(folder_path, train_file_names[i*2+1]))
    assert follower_file.shape[0] == leader_file.shape[0], "check the shape of the files"
    length = leader_file.shape[0]
    for j in range(0, length, 300):
        f_j = follower_file[j:j+300]
        l_j = leader_file[j:j+300]
        d_j = (f_j, l_j)
        train_result.append(d_j)
```

```python
for i in range(len(train_result)):
    length = train_result[i][0].shape[0]
    if length!=300:
        follower = np.zeros((300, 165))
        leader = np.zeros((300, 165))
        follower[:length] = train_result[i][0]
        leader[:length] = train_result[i][1]
        train_result[i] = (follower, leader)
```

```python
for data in train_result:
    if data[0].shape[0] != 300:
        print(data)
```
