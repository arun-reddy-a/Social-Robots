```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets
from ipywidgets import interact, IntSlider
import os
from torch.utils.data import Dataset, DataLoader
import torch.fft as fft
import torch
```

```python
train_data = np.load('data_more_processing/axis_magnitude_train_data.pkl', allow_pickle=True)
```

```python
sample_point = train_data[42][0]
```

```python
sample_point_angles = sample_point[:, 6:226].reshape(-1, 55, 4)
```

```python
sample_point_angles.shape
```

```python
sample_point_l_k = sample_point_angles[:, 4, :]
sample_point_r_k = sample_point_angles[:, 5, :]

sample_point_l_a = sample_point_angles[:, 7, :]
sample_point_r_a = sample_point_angles[:, 8, :]
```

```python
sample_point_l_k.shape
```

```python
def plot_freq_spectrum_update(data_point, str):
    '''data_point [300, 1]'''
    data_point = torch.from_numpy(data_point)

    fft_result = fft.rfft(data_point)
    amplitude = torch.abs(fft_result)

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(4, 5))
    ax1.plot(data_point.squeeze())
    ax1.set_title('Time Domain Signal '+str)
    ax1.set_xlabel('Time')
    ax1.set_ylabel('angle')
    
    ax2.plot(amplitude)
    ax2.set_title('Frequency Domain Signal')
    ax2.set_xlabel('Frequency')
    ax2.set_ylabel('Amplitude '+str)

    plt.title(str)
    plt.tight_layout()
    plt.grid()
    plt.show()
```

```python
point = sample_point_l_k
plot_freq_spectrum_update(point[:, -1], 'Left Knee Angle')
plot_freq_spectrum_update(point[:, 0], 'Left Knee x-axis')
plot_freq_spectrum_update(point[:, 1], 'Left Knee y-axis')
plot_freq_spectrum_update(point[:, 2], 'Left Knee z-axis')
```

```python
point = sample_point_r_k
plot_freq_spectrum_update(point[:, -1], 'Left Knee Angle')
plot_freq_spectrum_update(point[:, 0], 'Left Knee x-axis')
plot_freq_spectrum_update(point[:, 1], 'Left Knee y-axis')
plot_freq_spectrum_update(point[:, 2], 'Left Knee z-axis')
```

```python
point = sample_point_l_a
plot_freq_spectrum_update(point[:, -1], 'Left Knee Angle')
plot_freq_spectrum_update(point[:, 0], 'Left Knee x-axis')
plot_freq_spectrum_update(point[:, 1], 'Left Knee y-axis')
plot_freq_spectrum_update(point[:, 2], 'Left Knee z-axis')
```

```python
point = sample_point_r_a
plot_freq_spectrum_update(point[:, -1], 'Left Knee Angle')
plot_freq_spectrum_update(point[:, 0], 'Left Knee x-axis')
plot_freq_spectrum_update(point[:, 1], 'Left Knee y-axis')
plot_freq_spectrum_update(point[:, 2], 'Left Knee z-axis')
```
